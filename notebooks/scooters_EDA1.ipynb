```python
sql_null_query = '''
SELECT COUNT(*)
FROM scooters
WHERE pubdatetime IS NULL 
    OR latitude IS NULL
    OR longitude IS NULL
    OR sumdid IS NULL
    OR companyname IS NULL
;
'''

pd.read_sql(sql_null_query, con = engine)
```
